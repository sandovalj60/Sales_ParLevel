```python
import pandas as pd
```

```python

path_30to06 = "PLU_300523-060623.csv"
path_06to13 = "PLU_060623-130623.csv"
path_13to20 = "PLU_130623-200623.csv"


df_30to06 = pd.read_csv(path_30to06)
df_30to06.head(-50)
df_30to06.info()
# -------------------
# df_30to06.columns
```
